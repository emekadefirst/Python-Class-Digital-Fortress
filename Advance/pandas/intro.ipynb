```python
mortality = titanic["Survived"]
dead = []
alive =  []
for x in mortality:
    if x == 0:
        dead.append(x)

for y in mortality:
    if y == 1:
        alive.append(y)

total_dead = len(dead)
total_alive = len(alive)
total_mortality = total_alive + total_dead

perc_of_dead = total_dead/total_mortality * 100
perc_of_alive = total_alive/total_mortality * 100
print(f"Based on the analysis {perc_of_dead}% didn't survive the accident while {perc_of_alive}% survived")
```

```python
sales_data = pd.DataFrame(
    {
        "order ID": ['MOR1', 'MOR2', 'MOR3', 'MOR4'],
        "Product": ['Col-gate', 'Maclean', 'Oral-B', 'Pepsodent'],
        "Date": ['234', "2343", "453", "345"],
        "Price ($)": [123, 345, 567.77, 202.97]
    },
)
sales_data
```

```python
sales_data["Price (₦)"] = sales_data["Price ($)"] * 1660

sales_data
```

```python
new_sales_data = sales_data.rename(
    columns={
        "order iD": "order Number",
        "Product": "Item",
        "Price ($)": "Cost ($)",
        "Price (₦)": "Cost (₦)",
    }
)
new_sales_data
```

```python
mean = new_sales_data["Cost (₦)"]
mean.median()
```

```python
new_sales_data.agg(
    {
        "Cost (₦)": ["min", "max", "median", "skew"],
        "Cost ($)": ["min", "max", "median", "mean"],
    }
)
```

```python
new_sales_data[["Item", "Cost ($)"]].groupby("Item").mean()
```
